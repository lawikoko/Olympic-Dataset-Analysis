```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Load dataset

athletes = pd.read_csv('C:/Users/Abayomi/Downloads/archive (1)/athlete_events.csv')
regions = pd.read_csv('C:/Users/Abayomi/Downloads/archive (1)/noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

```python
#Join th edataset (Horizontal join)

athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')

athletes_df.head()
```

```python
athletes_df.shape
```

```python
#Renaming to make the column names consistent

athletes_df.rename(columns = {'region':'Region', 'notes':'Notes'}, inplace = True)
```

```python
#check null values

athletes_df.isnull().sum()
```

```python
#Print columns names containing null values or missing values in the form of a lsit

null_columns = athletes_df.columns[athletes_df.isnull().any()]

for i in null_columns:
    print(i)
```

```python
#Data for specific countries
#Nigeria

athletes_df.query('Team == "Nigeria"').head()
```

```python
#Top 10 paricipating countries

top_10_participants = athletes_df.Team.value_counts().sort_values(ascending = False).head(10)

top_10_participants
```

```python
#Plot for top 10 participating countries

plt.figure(figsize=(12,6))
plt.title('Overall Participation by Country')
sns.barplot(x = top_10_participants.index, y = top_10_participants, palette = 'Set2' );
```

```python
#Age Distribution of the Participants

plt.figure(figsize = (12, 6))
plt.title('Age distribution of Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'blue', edgecolor = 'white');
```

```python
#Winter Olympics sports

winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()

for i in winter_sports:
    print(i)
```

```python
#Summer Olympics sports

summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()

for i in summer_sports:
    print(i)
```

```python
#Male an dfemale participants

gender_counts = athletes_df.Sex.value_counts()

gender_counts
```

```python
#Pie plot for gender distribution

plt.figure(figsize = (12,6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels = gender_counts.index, autopct = '%1.1f%%', startangle = 0);
```

```python
#Total Medals

medal_counts = athletes_df.Medal.value_counts()

medal_counts
```

```python
#Total number of female participants in each olympics 

female_participants = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

```python
women_olympics = athletes_df[(athletes_df.Sex == 'F') & (athletes_df.Season == 'Summer')]
```

```python
#Creating a count plot

sns.set(style = 'darkgrid')
plt.figure(figsize = (20,10))
sns.countplot(x = 'Year', data = women_olympics, palette = 'Spectral')
plt.title('Women Participation')
```

```python
#Plotting a line graph for female participation

part = women_olympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize = (20,10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athlete Participation')
```

```python
#Gold medal athletes

gold_medal = athletes_df[(athletes_df.Medal == 'Gold')]
gold_medal
```

```python
#take only the values that are different from NaN

gold_medal = gold_medal[np.isfinite(gold_medal['Age'])]
```

```python
#Gold Medal beyong 60

gold_medal['ID'][gold_medal['Age'] > 60].count()
```

```python
#Checking the sports which the gold medalists participated in 

sporting_event = gold_medal['Sport'][gold_medal['Age']>60]
sporting_event.unique()
```

```python
#Plot for the sporting events

plt.figure(figsize = (10,5))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title('Sports Gold Medalists older 60 participated');
```

```python
#Gold medals from each country

gold_medal.Region.value_counts().reset_index(name = 'Medal').head()
```

```python
totaGoldMedals = gold_medal.Region.value_counts().reset_index(name = 'Medal').head(5)
g= sns.catplot(x = "index", y = 'Medal', data = totaGoldMedals, 
               height = 5, kind = 'bar', palette = 'rocket')

g.despine(left = True)
g.set_xlabels('Countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per Country')
```

```python
#Rio Olymoics

rio = athletes_df.Year.max()

team_names = athletes_df[(athletes_df.Year == rio) & (athletes_df.Medal == 'Gold')].Team

team_names.value_counts()
```

```python
sns.barplot(x = team_names.value_counts().head(20), y = team_names.value_counts().head(20).index)

plt.ylabel = (None);
plt.xlabel = ('Country Medals for the year 2016');
```

```python
#Height vs weight distribution

not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize = (12,10))
axis = sns.scatterplot(x = 'Height', y = 'Weight', data = not_null_medals, hue = 'Sex')
plt.title('Height vs Weight of Olympic Medalists')
```
